```python
import pandas as pd
import numpy as np


pd.set_option('display.max_columns',None)
```

```python
credit_df=pd.read_csv('credit_data.csv')
```

```python
credit_df.shape
```

```python
## Data preproseccing
```

```python
## Drop unimporttant columns

credit_df=credit_df.drop(['Customer_ID','Name','SSN','ID','Month'],axis=1)
```

```python
credit_df.shape
```

```python
credit_df.head()
```

```python
## Feature scaling

## Encoding categorical data

cat_data=credit_df[credit_df.columns[credit_df.dtypes=='O']]
num_data=credit_df[credit_df.columns[credit_df.dtypes!='O']]
```

```python
cat_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encode=LabelEncoder()
credit_df['Credit_Mix']=encode.fit_transform(credit_df['Credit_Mix'])
credit_df['Payment_of_Min_Amount']=encode.fit_transform(credit_df['Payment_of_Min_Amount'])
credit_df['Credit_Score']=encode.fit_transform(credit_df['Credit_Score'])
credit_df['Payment_Behaviour']=encode.fit_transform(credit_df['Payment_Behaviour'])
credit_df['Occupation']=encode.fit_transform(credit_df['Occupation'])
```

```python
credit_df['Occupation']
```

```python
num_features=num_data.columns
num_features
```

```python
## feature scalling for numerical features

from sklearn.preprocessing import MinMaxScaler

norm=MinMaxScaler()

credit_df[num_features]=norm.fit_transform(credit_df[num_features])

```

```python
credit_df
```

```python
X=credit_df.iloc[:,:-1]
Y=credit_df.iloc[:,-1]
```

```python
### features selection

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

fs=SelectKBest(score_func=f_regression,k=10)

X_best=fs.fit_transform(X,Y)

x=pd.DataFrame(X_best)
```

```python
data=pd.concat([x,Y],axis=1)
data.head()
```

```python
data.to_csv('credit_train_data.csv')
```
